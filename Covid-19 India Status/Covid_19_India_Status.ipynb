```python
plt.figure(figsize=(15,8))
ax = sns.heatmap(df.corr(),annot = True)
plt.show()
```

```python
x = df.drop(['State/UTs'],axis = 1)
for i in x.columns:
    plt.figure(figsize=(15,8))
    sns.histplot(x[i],kde = True)
    plt.show()
```

```python
x = df.drop('State/UTs',axis = 1)
for i in x.columns:
  plt.figure(figsize=(15,8))
  sns.boxplot(x = i, data = x,color = 'yellowgreen')   
  plt.xlabel(i)
  plt.show()
```

```python
x = df.drop('State/UTs',axis = 1)
for i in x.columns:
  plt.figure(figsize=(15,8))
  sns.violinplot(x = i, data = x)   
  plt.xlabel(i)
  plt.show()
```
